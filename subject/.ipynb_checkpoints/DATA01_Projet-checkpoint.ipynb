# Méthodologie de projet

## Introduction
Le projet de fin de session est destiné à faire une synthèse des différents points abordés pendant les cours.

## Méthode
### A. Définir la problématique

> Partie préliminaire de l'enquête. Chercher à poser une problématique, c'est-à-dire une question principale, éventuellement accompagnée de questions annexes ou intermédiaires, soutenue•s par des hypothèses que l‘on cherchera à conforter ou à invalider.
>
> La problématique présuppose (en théorie) une connaissance préalable du domaine de l‘étude, ce qui est évidemment difficile dans le cadre d'un exercice « scolaire ». Néanmoins, dans la mesure du possible, on cherchera à retracer l‘arrière-plan de la question principale.

### B. Préparation

#### 1. Identifier les sources de données / Expliciter la collecte
* Tâche
> Repérer les lister les données initiales 
* Livrable : **Etat des lieux des sources**
> Une liste des sources de données initialement utiles, comprenant :
> * L'organisme ayant publié la source
> * La localisation de la source
> * Le format de la source
> * La méthode de récupération de la source
> * Les problèmes éventuels d'accès ou de récupération de la source

#### 2. Comprendre les modèles des sources
* Tâche
> Examiner le•s jeu•x de données pour décrire les différents concepts dont ils rendent compte et qui seront utiles pendant l'enquête. Comprendre comment lesdonnées
* Livrable : **Modèle conceptuel**
> Des schémas du style « Entité-Association »,permettant de comprendre l'organisation conceptuelle des donnéees. Dans la mesure du possible, ces shémas retraceront les aspects “sémantiques” des associations et les contraintes s'exerçant sur elles.

#### 3. Examiner les données
* Tâche
> Examen approfondi des données dans leur relation avec le « modèle logique ». Recenser toutes les particularités et les défauts : types , manques, formes normales, mauvaises formes, incohérences, valeurs aberrantes, etc.
* Livrable : **Etat des lieux des données**
> Une synthèse des problèmes rencontrés sur ces données :
> * Une estimation de la qualité des données
> * Une liste des problèmes rencontrés
> * Les décisions prisespourpallier ces problèmes
> * La description d'un processus automatique de correction des données

#### 4. Produire un modèle de sortie pour l'enquête
* Tâche
> Sélectionner dans les donéees disponibles celles qui auront une importance dans les tratements ultérieurs.
* Livrable
> Une fois le modèle établi, on rendra :
> * Un schéma représentant le « modèle conceptuel »
> * Un schéma représentant le « modèle logique », c'est-à-dire le précédent converti en termes opérationnels.

#### 5. Aligner les modèles
* Tâche
> Comparer les(éventuelles) différentes sources pour harmoniser leurs descriptions en vue de l'enrichissement du modèle de sortie avec des données fiables.
* Livrable
> Un processus technique permettant d'harmoniser les sources vers le modèle de sortie.

### C. Processus technique
#### 1. Hypothèses sur les données
* Tâche
> Vérifier certaines hypothèses faites sur les données, par exemple en termes de distribution et d'indépendance
* Livrable
> Analyse des corrélations entre variables

#### 2. Choix d'une architecture technique
* Tâche
> En fonction de l'objet de l'enquête, choisir les outils qui permettront de mener des investigations dans le corpus de données : statistiques, analyse de texte, arbres de décision. C'est le moment de passer du « modèle logique » au « **modèle physique** » qui voit les objets de l'étude être convertis en objets proprment informatiques (fichiers, bases de données, programmes, etc.)
* Livrable
> A l'issue de cette étape,

#### 3. Ecriture des programmes
* Tâche
> Ecriture du code permettant l'automatisation du processus d'analyse de données. Cette phase est elle-même itérative et peut être exécutée sur des jeux de données de petite taille pour en vérifier la robustesse.
* Livrable
> Un package (en Python dans notre cas) synthétisant l'ensemble de tout le code nécessaire à l'exécution d l'étude et à sa répétition dans des conditions nouvelles

#### 4. Résultats
* Tâche
> Exécution du programme sur l'ensemble des données sélectionnées.
* Livrable
> Tous les éléments rendant possible l'interprétation comme réponse à la question posée. Ce peut être un ensemble de tableaux, de graphiques, d'expérimentations, de prédictions, de classifications, etc.

#### 5. Interprétation
* Tâche
> Analyse des résultats. Recherche d‘éventuels biais qui devraient être corrigés utltérieurement. Mise à jour d‘éléments manquants qui permettraient d'expliquer certains points de l'enquête. Etablissement de corrélations, voire de causalités, entre différentes variables.
* Livrable
> Une courte synthèse expliquant ce que vous tirez des résultats données par les programmes. Elle pourra éventuellement argumenter sur le fait qu'elle conforte ou non une (éventuelle) hypothèse préalable. Elle intégrera aussi les points restés indécidables.

### D. Présentation

> La phase ultime consite à préparer la présentation des résultats, qui dans notre cas est la soutenance.
>
> C'est une phase de communication (que nous n'avons pas développée). Préparez quelques planches, suffisamment synthétiques, qui vous permettront de retracer la problématique, la construction des données et les résultats.
>
> Vous pouvez vous baser sur a structure canonique des articles de recherche :
> * Abstract
> * Hypothèses
> * Matériel et méthodes
> * Conclusion et interprétation
> * Travail à venir
>
> Dans le cadre de la soutenance, on vous demandera naturellement de faire le lien avec ce qui a été vu pendant les cours.
